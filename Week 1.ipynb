```python
import numpy
import urllib
import scipy.optimize
import random
```

```python
def parseData(fname):
    for l in urllib.urlopen(fname):
        yield eval(l) 

# eval:将字符串string对象转化为有效的表达式参与求值运算返回计算结果,可以把list,tuple,dict和string相互转化。
          
print "Reading data..."
data = list(parseData("http://jmcauley.ucsd.edu/cse190/data/beer/beer_50000.json"))
print "done"
        
```

## Real-valued features

```python
data = [d for d in data if 'user/ageInSeconds' in d and d['user/ageInSeconds'] < 60*60*24*365*80]

def feature(datum):
    feat = [1,datum['user/ageInSeconds'],datum['user/ageInSeconds']**2]
    return feat

X = [feature(d) for d in data]
X[:10]
y = [d['review/overall'] for d in data]
y[:10]

theta,residuals,rank,s = numpy.linalg.lstsq(X, y)
#Return the least-squares solution to a linear matrix equation.
```

## Categorical features

```python
data2 = [d for d in data if d.has_key('user/gender')]

def feature(datum):
  feat = [1]
  if datum['user/gender'] == "Male":
    feat.append(0)
  else:
    feat.append(1)
  return feat

X = [feature(d) for d in data2]
y = [d['review/overall'] for d in data2]
theta,residuals,rank,s = numpy.linalg.lstsq(X, y)
```

## Random features

```python
def feature(datum):
  feat = [1] + [random.random() for i in range(100)]
  return feat

X =[feature(d) for d in data ]
X[:10]
```
